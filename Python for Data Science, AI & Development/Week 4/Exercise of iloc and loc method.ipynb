## Import Panda library and Selecting data in a Dataframe

```python
# import and select 5 row
import pandas as pd
df = pd.read_csv('employees.csv')
df.head()
```

```python
# loc works on row name and column name
# Here 0 is not an index, 0 is row name
df.loc[0, :] 
```

```python
# This will select specific row [0, 1, 3] using row name
df.loc[[0,1,3], :] # [0,1,3] - written as list
```

## Slicing

```python
# loc will select rows from 0 to 5 - total 6 rows

# Attention - 
# loc will consider 5 also [0, 1, 2, 3, 4, 5]
# iloc don't consider 5    [0, 1, 2, 3, 4]
df.loc[0:5, :]
```

```python
# Select only Gender column for first 6 rows
df.loc[0:5, 'First Name']
```

```python
# Select Gender and First Name column for first 6 rows
# put all column name in the list
df.loc[0:5, ['First Name', 'Gender']]
```

```python
# Select column from First Name to Salary for first 6 rows
df.loc[0:5, 'First Name':'Salary']
```

```python
# Print that condition is true or false for each row
#df['Gender']=='Female'

# Select all rows whowes gender is Female with only First Name column
# df.loc[df['Gender']=='Female','First Name']

# Select all rows whowes gender is Female with all column
df.loc[df['Gender']=='Female',:]
```
